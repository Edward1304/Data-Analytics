## **4. Trabajando con Datos en Notebooks**
- ### **Importacion de bibliotecas como NumPy y Pandas.**

```python
import pandas as pd
import numpy as np
```


- ###  **Carga y manipulacion basica de datos.**

```python
# Crear unos datos  como un arreglo con numpy que son  las tres temperaturas de tres dias 
# de la semana
data = np.array([[28, 34, 37], [30, 28, 34], [33, 33, 34]])

# convertir los datos a dataframe con pandas
df = pd.DataFrame(data, columns=['TempMon', 'TempTue', 'TempWen'])

# mostrar las columnas
print(df)

#muestra la  informacion de las columnas
df.apply(np.mean)

```

- ### **Visualizacion con Matplotlib.**



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,6))

# Grafico de lineas
plt.subplot(1, 2, 1)
plt.plot(df['TempMon'], label='Temperatura dia Lunes')
plt.plot(df['TempTue'], label='Temperatura dia Martes')
plt.plot(df['TempWen'], label='Temperaura dia Miercoles')
plt.title('Visualion de datos con matplotlib (Temperaturas de la semana)')
plt.xlabel('Index')
plt.ylabel('Grados Celsius')
plt.legend()

# Histograma
plt.subplot(1, 2, 2)
plt.hist(df.values.ravel(), bins=10, alpha=0.5, color='g', edgecolor='black')
plt.title('Histograma de las Temperaturas')
plt.xlabel('Grados Celsius')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

```

## **5. Ejemplo de Analisis de Datos**
- ### **Realización de un análisis simple utilizando los datos cargados.**


- ### **Generacion de graficos descriptivos.**

```python
# Cargas los datos
df_f1 = pd.read_csv('F1Drivers_Dataset.csv')

# Muestra las primeras 5 filas
print(df_f1.head())

# Define championships variable
championships = df_f1['Nationality'].value_counts()

# Set up the figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Histograma de las décadas respecto a los pilotos que han participado
axs[0, 0].hist(df_f1['Decade'], color='red')
axs[0, 0].set_title('Histograma de Decadas')
axs[0, 0].set_xlabel('Decada')
axs[0, 0].set_ylabel('Cantidad de Pilotos')

# Diagrama de dispersión de victorias y podios en carreras
axs[0, 1].scatter(df_f1['Race_Wins'], df_f1['Podiums'], color='violet')
axs[0, 1].set_title('Diagrama de dispersion Race_Wins vs Podiums')
axs[0, 1].set_xlabel('Race_Wins')
axs[0, 1].set_ylabel('Podiums')

# Gráfico circular para la distribución de los campeonatos.
axs[1, 0].pie(championships, labels=championships.index, autopct='%1.1f%%')
axs[1, 0].set_title('Distribution of Championships')

#Gráfico de barras para la distribución de las carreras ganadas por década.
axs[1, 1].bar(df_f1['Decade'], df_f1['Race_Wins'],color='green')
axs[1, 1].set_title('Carreas Ganadas por Decada')
axs[1, 1].set_xlabel ("Decada")
axs[1, 1].set_ylabel ('Race_Wins')


plt.tight_layout()
plt.show()
```
